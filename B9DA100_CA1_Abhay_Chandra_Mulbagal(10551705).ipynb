```
# TAX CALCULATIONS BASED ON THE MONTHLY TOTAL SALARY AND FINAL PAY AFTER DEDUCTION ######
# The Intent of this code based on Question 1A is to deduce in-hand salary on a month on month basis #####
# The user will be prompted to enter the details of an employee along with the pay this employee receives ###

# datetime library helps to validate date is set in the right format
from datetime import datetime

# Execution begins here when this function is invoked
def main():
    employee_pay_data = gather_worker_data()
    employee_pay_data = salary_post_deduction_calculator(employee_pay_data)
    # print(employee_pay_data)
    monthly_payslip_generator(employee_pay_data)


# This function performs calculations based on the input received 
# input: a dictionery with employee income data
# output: a dictionery with details on calculated salary after deductions
def salary_post_deduction_calculator(employee_income_data):

    employee_income_data['total_without_ot'] = round(
        employee_income_data['hours_worked'] * employee_income_data['hourly_rate'], 2)
    employee_income_data['standard_deduction'] = round(
        employee_income_data['total_without_ot'] * employee_income_data['standard_tax_rate'], 2)
    employee_income_data['total_after_ot'] = round(
        employee_income_data['overtime_worked'] * employee_income_data['overtime_rate'], 2)
    employee_income_data['ot_deduction'] = round(
        employee_income_data['total_after_ot'] * employee_income_data['overtime_tax_rate'], 2)
    employee_income_data['total_pay'] = round(
        employee_income_data['total_without_ot'] + employee_income_data['total_after_ot'], 2)
    employee_income_data['net_deduction'] = round(
        employee_income_data['standard_deduction'] + employee_income_data['ot_deduction'], 2)
    employee_income_data['net_pay'] = round(employee_income_data['total_pay'] - employee_income_data['net_deduction'],
                                            2)
    # print(employee_income_data)
    return employee_income_data


# The gather_worker_data() prompts user a number of inputs to provide salary
# details and tax information on the employee's income
# input: none
# output: A dictionery with user input values of employee's salary and tax data
def gather_worker_data():
    worker_data_dictionery = {}
    while True:
        worker_name = input("Enter the {} (Name should be alphabetical characters only) :".format("Employee Name"))
        if worker_name.isalpha():
            break
        else:
            continue
    worker_data_dictionery['employee_name'] = worker_name

    print()

    while True:
        worker_unique_id = input("Enter the {} (Use Alphanumeric characters only.) :".format("Employee Number"))
        if worker_unique_id.isalnum():
            break
        else:
            continue
    worker_data_dictionery['employee_id'] = worker_unique_id

    print()

    while True:
        try:
            worker_data_dictionery['week_ending'] = input(
                "Enter the week of the month when salary gets credited(in DD/MM/YYYY):")
            datetime.strptime(worker_data_dictionery['week_ending'], "%d/%m/%Y")
            break
        except ValueError:
            print("Please enter a valid date in the specified format")
    
    print()

    while True:
      try:
        worker_data_dictionery['hours_worked'] = float(input("The number of hours worked by the employee:"))
      except ValueError:
        print("Enter a numeric value only.")
        continue
      if isinstance(worker_data_dictionery['hours_worked'], float):
        break
      else:
        continue

    print()

    while True:
      try:
        worker_data_dictionery['hourly_rate'] = float(
            input("The amount to be paid for employee on per hour basis (Euros per hour):"))
      except ValueError:
        print("Enter a numeric value only.")
        continue
      if isinstance(worker_data_dictionery['hourly_rate'], float):
        break
      else:
        continue

    print()

    while True:
      try:
        worker_data_dictionery['overtime_worked'] = float(
            input("The additional hours of work performed by the employee:"))
      except ValueError:
        print("Enter a numeric value only.")
      if isinstance(worker_data_dictionery['overtime_worked'], float):
        break
      else:
        continue

    print()

    while True:
      try:
        worker_data_dictionery['overtime_rate'] = float(
            input("The amount to be paid for employee for overtime work on hourly basis:"))
      except ValueError:
        print("Enter a numeric value only.")
      if isinstance(worker_data_dictionery['overtime_rate'], float):
        break
      else:
        continue

    print()

    while True:
      try:
        tax_in_percent = float(
            input("Enter the standard tax rate on the income in percent (slab level of 20% 0r 30% so on):"))
      except:
        print("Enter numeric value only.")
      if isinstance(tax_in_percent, float):
        break
      else:
        continue

    worker_data_dictionery['tax_in_percent'] = tax_in_percent
    tax_in_percent = tax_in_percent / 100
    worker_data_dictionery['standard_tax_rate'] = tax_in_percent

    print()

    while True:
      try:
        ot_tax_in_percent = float(input("Enter the over-time tax rate on the income in percent ( 20% or 30% so on):"))
      except ValueError:
        print("Enter a numeric value only.")  
      if isinstance(ot_tax_in_percent, float):
        break
      else:
        continue

    worker_data_dictionery['ot_tax_in_percent'] = ot_tax_in_percent
    ot_tax_in_percent = ot_tax_in_percent / 100
    worker_data_dictionery['overtime_tax_rate'] = ot_tax_in_percent

    # print(worker_data_dictionery)
    return worker_data_dictionery


# The monthly_payslip_generator function uses the processed user data after
# tax deduction and generates formatted payslip
# input: dictionery object containing deductions processed data
# output: none (prints formatted payslip)
def monthly_payslip_generator(final_payslip):
    print()
    print("{:>90}".format("PAYSLIP"))
    print("*" * 180)
    print("{:>50} WEEK ENDING {}".format('', final_payslip['week_ending']))
    print("{:>50} Employee Name: {}".format('', final_payslip['employee_name']))
    print("{:^50} Employee Number: {}".format('', final_payslip["employee_id"]))
    print("{:>80}{:>30}".format('Earnings', "Deductions"))
    print("{:>77}{:>6}{:>10}".format('Hours', "Rate", "Total"))
    print(
        "{:>56}{:>8}{:>13}{:>6}{:>10}{:>10}{:>2}{:>5}% {:>6}".format('Hours', "(normal)", final_payslip['hours_worked'],
                                                                     final_payslip['hourly_rate'],
                                                                     final_payslip['total_without_ot'], "tax", "@",
                                                                     final_payslip['tax_in_percent'],
                                                                     final_payslip['standard_deduction']))
    print("{:>56}{:>8}{:>11}{:>6}{:>10}{:>10}{:>2}{:>5}% {:>6}".format('Hours', "(overtime)",
                                                                       final_payslip['overtime_worked'],
                                                                       final_payslip['overtime_rate'],
                                                                       final_payslip['total_after_ot'], "tax", "@",
                                                                       final_payslip['ot_tax_in_percent'],
                                                                       final_payslip['ot_deduction']))
    print()
    print("{:>82}{:>34}".format("Total Pay:", final_payslip['total_pay']))
    print("{:>89}{:>27}".format("Total Deductions:", final_payslip['net_deduction']))
    print("{:>82}{:>34}".format("Net Pay:", final_payslip['net_pay']))
    print("*" * 180)


if __name__ == '__main__':
    main()
```

```
# Solution 3: This solution for problem 3 which requires users to provide integers
# store them and count the frequency of each element, makes use of in built library
# collections that provides a counter class object which takes a list and calculates 
# list individual elements frequencies

# The counter class object is imported for counting list elements
from collections import  Counter
user_entries_accumulator=[]


# Exceution starts here through main()
def main():
  #count=0
  accumulator_list_size=int(input("Input the number of elements to be stored in the list:"))
  print("Input {} elements to the list".format(accumulator_list_size))
  for iter in range(accumulator_list_size):
    val=int(input("element-{} : ".format(iter)))
    user_entries_accumulator.append(val)
  print(user_entries_accumulator)

  # The counter object counts and includes repeated values as tuples in 
  # to repeated_numbers_accumulator
  repeated_numbers_accumulator=Counter(user_entries_accumulator)
  print("The frequency of all elements in4 the list: ")
  for key,value in repeated_numbers_accumulator.items():
    print("{} occurs {} times".format(key,value))


if __name__ == '__main__':
    main()
```

```
# Solution 4: For the problem 4, The Phonebook app incorporates database concepts
# learnt in class and to perform create, retrieve, update and delete operations (CRUD),
# database sqlite 3 is used to have a hands-on experience. This program also
# incorporates few error handling scenarios required for smooth database operations
# and connection creation and closures

# sqlite3 library is used for required add, remove, update and search ops
# as database service provider
import time as timer
import sqlite3

# The database connection function creates exactly one instance of database object
# to be used per app invocation. 
# This app can be greatly improved by incorporating OOP concepts and avoid 
# deadlock issues etc.

def database_connection():
  db2 = sqlite3.connect("MYPY_Phonebook.db")
  db2.execute("drop table if exists results")
  try:
    db2.execute("create table phonebook(full_name TEXT, phone_number INTEGER NOT NULL UNIQUE)")
    print("Please wait for few seconds until database is created....")
    timer.sleep(20)
  except:
    print("Already table exists !!")
  return db2

# Execution begins here from main()
def main():

  print("#" * 60)
  print("MYPY PHONE BOOK")
  print("#" * 60)

# create a database instance to connect and perform CRUD operations
  db2 = database_connection()

  while True:
    try:
      user_selection=int(input("""
      1: Add New Entry
      2: Delete Entry
      3: Update Entry
      4: Lookup Number
      5: QUIT
      """))
    except ValueError:
      print(" Please enter a valid option ranging from 1 to 5.")
      continue


    # This option leads to create new user and phone number rows in database
    if user_selection==1:
      phone_user_name=input("Enter the User's Full Name:")
      user_phone_number=int(input("Enter {}'s Phone Number:".format(phone_user_name)))
      check_processing=add_user_phone_number(db2,phone_user_name,user_phone_number)
      if check_processing:
        print("The number successfully added!")
      else:
        print("Duplicate entry/invalid details were provided!")

     
    # This option is used to remove information from the database 
    elif user_selection==2:
      phone_number=int(input("Enter the Phone number of the user to be deleted from records:"))
      ## check before removal if number is present - can use lookup set a flag to provide or not provide add entry option
      update_required=False
      remove_phone_number(db2,phone_number,update_required)
      #print("The {} number's record has been removed from the phonebook".format(phone_number))


    # This option is used to update user information based on phonenumber
    elif user_selection==3:
      old_number=int(input("Enter the existing phone number to be modify or update:"))
      ##code to check  if number is present --> lookup and add
      modified_number=int(input("Enter the new phone number:"))
      data_add_flag=False
      modify_entry(db2,old_number, modified_number,data_add_flag)
      # print("This number {} has been updated and added to the phonebook.".format(modified_number))

       
    # This option used to search a number based on phone number
    elif user_selection==4:
      lookup_phone_number=int(input("Enter a phone number to search:"))
      add_new_entry=False
      data=search_entry(db2, lookup_phone_number, add_new_entry)


    # This option terminates the execution but also ensures database object is
    # closed safely.
    elif user_selection==5:
      print("Thank you for using the MYPY Phonebook Application! Stay safe! Stay healthy!")
      db2.close()
      break

    else:
      print("That was not a valid input. Please enter the input as specified.")
      continue


# "add_user_phone_number" function uses user inputs and inserts this data to
# to the database. It also returns a successfully_added flag to indicate if the 
# the data insert operation was successful or not. This return value helps in
# search operations to include add if data not present
# inputs: database object, (string)user_name_new, (int)phone_number_new
# output: (boolean) successfully_added

def add_user_phone_number(db2, user_name_new, phone_number_new):
  successfully_added = True
  try:
    # INSERT cmd to insert user and phone number into the database
    cmd = "insert into phonebook(full_name, phone_number) values('{}',{})".format(user_name_new, phone_number_new)
    db2.execute(cmd)
    successfully_added = True
    db2.commit()
  except:
    print("This number already exists! Please enter another number.")
    successfully_added = False
    db2.commit()
  return successfully_added


# The remove_phone_number function removes data from phonebook database based on
# phone_number. The lookup_update_flag is used to search and indicate if data is
# present to delete after searching and retrieving the data
# input(s): database object, (int) phone_number, (boolean) lookup_and_update_flag
# output: None (prints if data isnt available in database to delete)
def remove_phone_number(db2,phone_number, lookup_and_update_flag):
  cur=db2.cursor()
  data=search_entry(db2, phone_number, lookup_and_update_flag)
  if data:
    # DELETE command query to remove a row based on phone number
    cmd="DELETE FROM phonebook WHERE phone_number={}".format(phone_number)
    db2.execute(cmd)
    db2.commit()
    print("The number has been deleted.")
  else:
    print("The number isn't available in the phonebook to delete.")


# modify_entry function updates the database based on phone number.
# input(s): database object, (int) phone_number_current, (int) phone_number_modify,
# (boolean) check_data (this flag is used for searching the phone number for existence before update) 
# output: None (Prints number isnt present to update if update query returns null)
def modify_entry(db2, phone_number_current, phone_number_modify, check_data):
  data=search_entry(db2, phone_number_current, check_data)
  cur=db2.cursor()
  if data:
    # UPDATE command to reset phone number of an existing phone number
    cmd="UPDATE phonebook SET phone_number = {} WHERE phone_number={}".format(phone_number_modify,phone_number_current)
    db2.execute(cmd)
    db2.commit()
    print("The number {} has been updated to {}.".format(phone_number_current,phone_number_modify))
  else:
    print("The number isn't available in the phonebook to update.")


# search_entry function searches the database based on the phone number.
# This function is very useful in add, remove, update operations as well as it 
# ensures data existence before perform any operation
# input(s): database object, (int) search_phone_number, (boolean) add_entry
# output(s): a tuple object containing row-wise data of the searched number.

def search_entry(db2,search_phone_number,add_entry):
  # The cursor objects enables the app to execute queries and perform detailed
  # operations based on fetched rows/columns
  cur=db2.cursor()
  # SELECT command to search user and phone number based on phone_number
  cmd="SELECT full_name, phone_number from phonebook WHERE phone_number={}".format(search_phone_number)
  data=cur.execute(cmd)
  rows=cur.fetchall()
  if rows:
    for row in rows:
      print("The existing/Searched Number Details:",row)
  else:
    print("Unable to find the records for this number.")
    user_value=input("Enter Y to add this number to the phonebook. Enter N to return to main menu :")
    add_entry= True if user_value == 'Y' else False
    if add_entry:
      user_name_update_new=input("Enter the user name:")
      add_entry=add_user_phone_number(db2,user_name_update_new, search_phone_number)
      print("This number has been added to the database!")

  db2.commit()
  return rows


if __name__ == '__main__':
    main()
```
